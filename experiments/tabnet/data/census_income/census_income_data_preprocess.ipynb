```python
import torch
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import train_test_split

import pandas as pd
import numpy as np
np.random.seed(0)

import os
import wget
from pathlib import Path

from matplotlib import pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("data.csv", names=list(range(15)))
```

```python
data.head()
```

```python
data["TARGET"] = data[14]
data = data.drop([14], axis=1)
```

```python
categorical_columns = [1, 3, 5, 6, 7, 8, 9, 13]
```

```python
data["TARGET"] = LabelEncoder().fit_transform(data["TARGET"])
```

```python
features = [col for col in categorical_columns] 
for feat in features:
    data[feat] = LabelEncoder().fit_transform(data[feat])
```

```python
data.head()
```

```python
data.to_csv("train.csv", index=False)
```

```python
data = pd.read_csv("train.csv")
```

```python
train_ratio = 0.70
validation_ratio = 0.15
test_ratio = 0.15
```

```python
train, test = train_test_split(data, test_size=1 - train_ratio, random_state=42, shuffle=True)
valid, test = train_test_split(test, test_size=test_ratio/(test_ratio + validation_ratio), random_state=42, shuffle=True)
```

```python
data.to_csv("data.csv", index=False)
```

```python
train.to_csv("train.csv", index=False)
valid.to_csv("val.csv", index=False)
test.to_csv("test.csv", index=False)
```
